```python
''' Write a Python program to create a list, a dictionary, and a set. 
Perform basic operations like adding, removing, and modifying elements. '''
```

```python
# LIST OPERATIONS
def list_creation():
    return [1,2,3,4]

def adding_to_list(a_list,element):
    a_list.append(element)

def removing_from_list(a_list,element):
    a_list.remove(element)

def modify_list(a_list,index,value):
    a_list[index] = value
    
def list_operation():
    a_list = list_creation()
    print("CREATED LIST: ",a_list)
    
    adding_to_list(a_list,5)
    print("LIST AFTER ADDING ELEMENT 5: ",a_list)
    
    removing_from_list(a_list,3)
    print("LIST AFTER REMOVING ELEMENT 3: ",a_list)
    
    modify_list(a_list,2,8)
    print("LIST AFTER MODIFYING INDEX 2: ",a_list)


# DICTIONARY OPERATIONS
def dict_creation():
    return {"a": 1, "b": 2, "c": 3}

def adding_to_dict(my_dict, key, value):
    my_dict[key] = value

def removing_from_dict(my_dict, key):
    del my_dict[key]

def modify_dict(my_dict, key, new_value):
    my_dict[key] = new_value

def dictionary_operation():
    my_dict = dict_creation()
    print("CREATED DICTIONARY: ",my_dict)

    adding_to_dict(my_dict, "d", 4)
    print("DICTIONARY AFTER ADDING KEY 'd' WITH VALUE 4: ",my_dict)

    removing_from_dict(my_dict, "b")
    print("DICTIONARY AFTER REMOVING KEY 'b': ",my_dict)

    modify_dict(my_dict, "a", 8)
    print("DICTIONARY AFTER MODIFYING VALUE OF KEY 'a' to 10: ",my_dict)


# SET OPERATIONS -- Modification of values in set is not possible as order of set changes everytime it is used.
def set_creation():
    return {2, 3, "apple"} #set with different datatype

def adding_to_set(my_set, element):
    my_set.add(element)

def removing_from_set(my_set, element):
    my_set.remove(element)

def set_operation():
    my_set = set_creation()
    print("CREATED SET: ",my_set)

    adding_to_set(my_set, 4)
    print("SET AFTER ADDING 4: ",my_set)

    removing_from_set(my_set, 2)
    print("SET AFTER REMOVING 2: ",my_set)


def main():
    print("List Operations")
    list_operation()

    print("\nDictionary Operations")
    dictionary_operation()

    print("\nSet Operations")
    set_operation()

if __name__ == "__main__":
    main()
```
